```python
import pyCalc as p
import matplotlib.pyplot as plt
```

```python
a, b, x = p.var('i'),p.var('i'), p.var()
x.f(a,b,'+')
sig = p.var()
sig.sigmoid(x)
xx = [i/10 for i in range(-40,50)]
ys = [sig.value([i, 0]) for i in xx]
plt.plot(xx,ys)
plt.title('Sigmoid')
plt.show()
print(sig.getDeriv(0,[0]))
```

```python
xx = [i/10 for i in range(-40,50)]
ys = [sig.getDeriv(0,[i, 0]) for i in xx]
plt.plot(xx,ys)
plt.title('Sigmoid deriv')
plt.show()
```
